```python
import pandas as pd
```

#Load Data

```python
df = pd.read_csv("E:/heatmap bharat prototype/heatmap-bharat-prototype/backend/data/city_day.csv")
```

# Keep Required Columns

```python
df = df[["City", "Date", "AQI"]]
```

# Remove Missing AQI

```python
df = df.dropna(subset=["AQI"])
```

# City Coordinates mapping

```python
city_coords = {
    "Delhi": (28.61, 77.23),
    "Mumbai": (19.07, 72.87),
    "Ahmedabad": (23.03, 72.58),
    "Bengaluru": (12.97, 77.59),
    "Chennai": (13.08, 80.27),
    "Kolkata": (22.57, 88.36),
    "Hyderabad": (17.38, 78.48),
    "Pune": (18.52, 73.85)
}
```

# Add latitude and longitude 

```python
df["lat"] = df["City"].map(lambda x: city_coords.get(x, (None, None))[0])
df["lon"] = df["City"].map(lambda x: city_coords.get(x, (None, None))[1])
```

# Normalize AQI

```python
df["intensity"] = df["AQI"] / df["AQI"].max()
```

# Save cleaned file

```python
df.to_csv("processed/clean_data.csv", index=False)

print(" Data processed successfully")
```
